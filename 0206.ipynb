### 학습용/검증용 데이터 분리

```python
# train set, test set으로 분리
#import sklearn.model_selection as ms
# train_test_split(arrays, test_size, random_state,shuffle) -> X_train, X_test, Y_train, Y_test = train_test_split(iris.data, test_size=0.3, random_state=42)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.3, random_state = 42) 
```

```python
len(X_train)
```

### 모델 구축

```python
# 1) 분석을 수행하기 위한 추정자(estimator) 생성
# 2) 학습 수행: 생성된 객체에 대해 fit() 메소드 수행
# 3) 예측 결과 도출: 생성된 객체 또는 학습이 수행된 분석 결과에 대해 predict()메소드 수행
```

```python
from sklearn.tree import DecisionTreeClassifier
# 1) 분석을 수행하기 위한 추정자(estimator) 생성자
cls = DecisionTreeClassifier()
# 2) 학습 수행: 생성된 객체에 대해 fit() 메소드를 수행
cls.fit(X_train,y_train)
# 3) 예측 결과 수행
y_pred = cls.predict(X_test)
y_pred
```

### 모델 평가

```python
# 생성된 객체 또는 학습이 수행된 분석 결과에 대해 적합한 성능 평가 지표 도출
# from sklearn.metrics import ****
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

- 원-핫 인코딩(One-Hot encoding)
- 해당 고유값만 1, 나머지는 0

```python
# 먼저 라벨인코딩으로 변환 후 원핫인코딩
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import numpy as np
items = ['Tv','냉장고','전자렌지', '컴퓨터', '선풍기',' 선풍기', '믹서', '믹서']
# 라벨 인코딩
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
# 원,핫 인코딩으로 변환하기 전에 2차원 배열 구조로 변경
labels = labels.reshape((-1,1)) # 값이 저장되지는 않음 (2,-1)
# 원핫인코딩
oh_encoder = OneHotEncoder()
oh_encoder.fit(labels)
oh_labels = oh_encoder.transform(labels)
oh_labels.toarray()
```

```python
# 원본 값 확인
oh_encoder.inverse_transform(oh_labels)
```

- 데이터 스케일링(피처 스케일링): 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업, 실제 특성값의 크기와 상관없이 동일한 조건으로 비교
- 표준화(standardization): 평균 0, 분산 1인 가우시안 정규 분포를 가진 값으로 변화 / x_new = (x-mean(x))/stdev(x)-표준편차
- 정규화(normalization): 서로 다른 피처의 크기를 통일하기 위해 변환(0~1)

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
iris.data
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
np.set_printoptions(precision =6, suppress=True)
iris_scaled_df = pd.DataFrame(iris_scaled)
pd.options.display.float_format = '{:.5f}'.format
iris_scaled_df.describe()
```

- MinMaxScaler()

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(iris.data)
scaler.transform(iris.data)
```

- fit() 데이터 변환을 위한 기준 정보 설정
- transform() 설정한 정보로 데이터를 변환
- fit_transform() fit과 transform을 한 번에 적용 및 수행
## 학습 데이터 세트와 테스트 데이터 세트에 fit(), transform() 적용 시 유의사항
- 가능하다면 전체 데이터의 스케일링 변환을 적용한 뒤 학습, 데이터 세트로 분리
- 1항이 안되면 테스트 데이터 변환 시에 fit()이나 fit_transform()을 적용하지 않고 학습 데이터로 이미 fit()된 Scaler객체를 사용해 transform()으로 변환
- Scaler()객체를 이용해 학습 데이터 세트로 fit, transform으로 한 것에는 다시 fit하면 안됨. 바로 transform 해야 함
